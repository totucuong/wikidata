Weekly graph specificity is defined as:

$$specificy = \frac{\sum_{a\in artifacts}{dist(a,root)}}{|artifacts|}$$, where an artifact is an Wikidata's item or property.

We plot specificity over time.

```python
def readData(pathToFile):
    f = open(pathToFile, 'r')
    l = f.read()
    dists = l.split(";")
    distV = [float(x) for x in dists]
    return distV

weeklySpecs = readData("../data/distance/wtp-distance")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

timeSeries = {}
timeSeries['week'] = [x for x in range(1, len(weeklySpecs) + 1)]
timeSeries['specificity'] = weeklySpecs
df = pd.DataFrame(timeSeries, index =timeSeries['week'])
df.plot(y='specificity', x='week')
plt.show()
```
